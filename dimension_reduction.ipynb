# Qualitative Comparison of Multiple Algorithms

```python
import matplotlib.pyplot as plot
import umap
from sklearn import decomposition, manifold
from sklearn.datasets import fetch_openml, load_digits
```

```python
algorithms = [
    ("UMAP", umap.UMAP(n_components=2)),
    ("tSNE", manifold.TSNE(n_components=2)),
    ("SVD", decomposition.TruncatedSVD(n_components=2)),
    ("PCA", decomposition.PCA(n_components=2))
]

fmnist, fmnist_labels = fetch_openml(data_id=40996, return_X_y=True)
digits, digits_labels = load_digits(return_X_y=True)
shuttle, shuttle_labels = fetch_openml(data_id=40685, return_X_y=True)
mnist, mnist_labels = fetch_openml(data_id=554, return_X_y=True)

def show_res(dataset, labels, dataset_name):
    fig, axes = plot.subplots(1, 4, figsize= (16,4))
    fig.suptitle(dataset_name, fontsize=20)
    for id, (name, algorithm) in enumerate(algorithms):
        data = algorithm.fit_transform(X=dataset)

        axes[id].set_xlabel(name)
        axes[id].scatter(data[:,0], data[:, 1], c=labels.astype(int))

```

```python
show_res(fmnist, fmnist_labels, "fmnist")
```

```python
show_res(digits, digits_labels, "digits")
```

```python
show_res(shuttle, shuttle_labels, "shuttle")
```

```python
show_res(mnist, mnist_labels, "mnist")
```
